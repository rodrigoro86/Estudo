As plotagens na matplotlib ficam em um objeto Figure

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 4)
```

```python
ax1.plot([1.5, 3.5, -2, 1.6], 'k--')
```

```python
ax3.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
```

```python
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)
axes
#share faz com que a escala dos gráficos sejem os mesmos
```

O array de axes pode ser facilmente indexado como um array bidimensional, por exemplo, axes[0, 1]

```python
axes[1, 1].plot(np.random.randn(50).cumsum(), 'k--')
```

Ajustando o espaçamento em torno das subplotagens

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

Cores, Marcadores e Estilos de Linha

```python
fig, axes = plt.subplots(1, 1)
axes = plt.plot(np.random.randn(30).cumsum(),'go--')
```

```python
fig, axes = plt.subplots(1, 1)
axes = plt.plot(np.random.randn(30).cumsum(), color='r', linestyle='dashed', marker='o')
```

```python
fig, axes = plt.subplots(1, 1)
data = np.random.randn(30).cumsum()
data
axes = plt.plot(data, color='r', linestyle='dashed', marker='o', label='Default')
```

```python
axes = plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post') #Faz a linhas pretas
```

```python
plt.legend(loc='best')#Cria a Legenda
```

Definindo o título, os rótulos dos eixos, os tiques e os rótulos dos tiques

Acrescentando legendas

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
```

```python
ax.plot(np.random.randn(50).cumsum(), 'k', label='one')
ax.plot(np.random.randn(50).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(50).cumsum(), 'k.', label='_nolegend_')
```

```python
ax.legend(loc='best')
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

myDates = [datetime(2012,1,i+3) for i in range(10)]
myValues = [5,6,4,3,7,8,1,2,5,4]
fig, ax = plt.subplots()
ax.plot(myDates,myValues)

myFmt = DateFormatter("%m-%d")
ax.xaxis.set_major_formatter(myFmt)

## Rotate date labels automatically
fig.autofmt_xdate()
plt.show()
```
